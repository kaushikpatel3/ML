```python
x=data.iloc[:,3:].values
x
```

```python
kmeans=KMeans(n_clusters=4,random_state=0)
kmeans.fit(x)
```

```python
kmeans.fit_predict(x)
```

```python
centre=kmeans.cluster_centers_
centre
```

```python
kmeans.inertia_
```

```python
labels=kmeans.labels_
labels
```

```python
plt.scatter(data["Annual Income (k$)"], data["Spending Score (1-100)"],c=labels)
```

```python
inertias=[]
for i in range(1,10):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(x)
    inertias.append(kmeans.inertia_)
    print("Inertia Values:",kmeans.inertia_)
    
plt.plot(range(1,10), inertias, marker='o')
plt.title("Elbow Method")
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 5, random_state = 0)
kmeans.fit(x)
```

```python
kmeans.inertia_
```

```python
plt.scatter(data["Annual Income (k$)"], data["Spending Score (1-100)"], c=labels)
plt.scatter(centre1[:, 0],centre1[:, 1], marker='*',color='r',linewidths=2)
```

```python
fit=kmeans.fit_predict(x)
plt.scatter(x[fit == 0,0] , x[fit == 0,1] , s=100 , c='darkviolet',label='cluster 0')
plt.scatter(x[fit ==1,0], x[fit==1,1],s=100, c='blue', label='cluster 1')
plt.scatter(x[fit ==2,0], x[fit==2,1],s=100, c='green', label='cluster 2')
plt.scatter(x[fit ==3,0], x[fit==3,1],s=100, c='lightgreen', label='cluster 3')
plt.scatter(x[fit ==4,0], x[fit==4,1],s=100, c='yellow', label='cluster 4')

plt.scatter(centre1[:, 0] , centre1[:, 1], s=200, marker= '*' ,color='k')

plt.legend()
plt.show()
```
